## Exploratory Data Analysis üìä üìà üìâ

##### Ratios General Function

```python
def plot_ratios(col, fig_text, feature_names=False, fig_title=None):
    features = col.value_counts().keys().to_numpy()
    show_status = np.array(["Show", "No Show"])

    plt.rc('font', size=12)
    fig, ax = plt.subplots(1,len(features), figsize=(16, 4))
    
    plt.figtext(0.5, 1, fig_text, ha="center", fontsize=18)

    for i, feature in enumerate(features):
        if i == 1: fig.legend()
        feature_showed_prob = len(dataset[(col == feature) & (dataset["No-show"] == "No")])/len(dataset[(col == feature)]) * 100
        did_not_show_ratio = 100 - feature_showed_prob
        counts = np.array([feature_showed_prob, did_not_show_ratio])

        ax[i].axis('equal')
        ax[i].pie(counts, labels=show_status, autopct='%1.2f%%') 
        if feature_names is False:
            ax[i].set_title(feature)
        else:
            ax[i].set_title(str(col.name) +" "+ str(col.value_counts().keys()[i]))
        ax[i].text(0.2, -1.8, round(feature_showed_prob-showup_probability,2), ha="center", fontsize=18)
```

## Q1: Which -from the given chartistics- effect the patients show up the most?

```python
# Appointments Vs. showing up status
labels = dataset["No-show"].value_counts().keys().to_numpy()
y = dataset["No-show"].value_counts().to_numpy()

plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(labels, y)
ax.set_ylabel("Appointments")
ax.set_xlabel("No-show")
ax.grid(True)    
ax.set_title("Appointments Vs. No-show")

# probability of patient showing up!
showup_probability = round((len(dataset[dataset["No-show"] == "No"])/ dataset.shape[0])*100,3) 

plt.figtext(0.5, 0, "{} of people showed while {} did not show up.".format(showup_probability, round(100-showup_probability,3)), ha="center", fontsize=18)
plt.show()
```

#### Column (1 of 11) : Gender Vs. No show üìä

### Now we will see what column has a different distribution from this overall distribution

```python
plt.rc('font', size=12)
fig, ax2 = plt.subplots(figsize=(10, 6))

# stack bar male/female appointments vs. showing up plot
labels = dataset.loc[dataset["Gender"].index, "No-show"].value_counts().keys().to_numpy()
male = dataset.loc[dataset["Gender"] == "M", "No-show"].value_counts().to_numpy()
female = dataset.loc[dataset["Gender"] == "F", "No-show"].value_counts().to_numpy()
ax2.bar(labels, male, label='Male')
ax2.bar(labels, female, bottom=male,label='Female')
ax2.set_ylabel('Appointments')
ax2.set_title('Appointments Vs. Male/Female No-show')
ax2.legend()
ax2.grid(True)

# statistical observation
ratio_of_male_show = len(dataset[(dataset["Gender"] == "M") & (dataset["No-show"] == "No")])/ len(dataset[dataset["Gender"] == "M"])
ratio_of_female_show = 1 - ratio_of_male_show

plt.figtext(0.5, 0.02, "{}% of males show at their appointment while females are {}%".format(round(ratio_of_male_show*100,3), round(ratio_of_female_show*100,3)), ha="center", fontsize=18)
plt.show()
```

> Observation: It's clear that gender has no effect on whether a patient will show or not.

#### Column (4 of 11) : Neighbourhood Vs. No show üìä

```python
print("There are {} Hospitals in the dataset".format(len(dataset["Neighbourhood"].value_counts().keys().to_numpy())))
```

```python
hospitals = np.sort(dataset["Neighbourhood"].value_counts().keys().to_numpy())
show_status = np.array(["Show", "No-show"])

hospital_showed_prob_array=np.zeros((len(hospitals)))
for i, hospital in enumerate(hospitals):
    hospital_showed_prob_array[i] = len(dataset[(dataset["Neighbourhood"] == hospital) & (dataset["No-show"] == "No")])/len(dataset[(dataset["Neighbourhood"] == hospital)]) * 100
hospital_name_array = np.array([hospital for hospital in hospitals])

# Top 5 hospitals that increase the prob of showing up
ind = hospital_showed_prob_array.argsort()[-5:][::-1]
top_5_ratios = hospital_showed_prob_array[ind]
top_5_hospitals = hospital_name_array[ind]

plt.rc('font', size=12)
fig, ax = plt.subplots(1,5, figsize=(16, 4))

for i, hospital in enumerate(top_5_hospitals):
    if i == 1: fig.legend()
    hospital_showed_prob = len(dataset[(dataset["Neighbourhood"] == hospital) & (dataset["No-show"] == "No")])/len(dataset[(dataset["Neighbourhood"] == hospital)]) * 100
    did_not_show_ratio = 100 - hospital_showed_prob
    counts = np.array([hospital_showed_prob, did_not_show_ratio])
    
    ax[i].axis('equal')
    ax[i].pie(counts, labels=show_status, autopct='%1.2f%%') 
    ax[i].set_title(hospital)
    ax[i].text(0.2, -1.8, round(hospital_showed_prob-showup_probability,2), ha="center", fontsize=18)
```

Observations: Location of hospitals have the heighest effect on the patients' show up so far.
<ul>
    <li>At PARQUE INDUSTRIAL neighbourhood all patients showed up. Chance is increased by 20%!</li>
</ul>

##### Describe General Function

```python
# quick and general 1d describe function
def describe_variable(col, xlabel, ylabel, title):
    scholarship_count = col.value_counts()
    ax = scholarship_count.plot(kind='bar')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
```

```python
describe_variable(dataset["Scholarship"], xlabel="Scholarship", ylabel="Appointments", title="Appointments Vs Scholarship count")
```

```python
plot_ratios(dataset["Scholarship"], fig_text="How Scholarship effect the probability of a patient to show up", feature_names = True)
```

```python
describe_variable(dataset["Hipertension"], xlabel="Hipertension", ylabel="Appointments", title="Appointments Vs Hipertension count")
```

```python
plot_ratios(dataset["Hipertension"], fig_text="How Hipertension effect the probability of a patient to show up", feature_names = True)
```

```python
describe_variable(dataset["Diabetes"], xlabel="Diabetes", ylabel="Appointments", title="Appointments Vs Diabetes count")
```

```python
plot_ratios(dataset["Diabetes"], fig_text="How Diabetes effect the probability of a patient to show up", feature_names = True)
```

```python
describe_variable(dataset["Alcoholism"], xlabel="Alcoholism", ylabel="Appointments", title="Appointments Vs Alcoholism count")
```

```python
plot_ratios(dataset["Alcoholism"], fig_text="How Alcoholism effect the probability of a patient to show up", feature_names = True)
```

```python
describe_variable(dataset["Handcap"], xlabel="Handcap", ylabel="Appointments", title="Appointments Vs Handcap count")
```

```python
plot_ratios(dataset["Handcap"], fig_text="How Handcap effect the probability of a patient to show up", feature_names = True)
```

Observations:
